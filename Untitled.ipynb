```python
elzamalek_fft = rfft(elzamalek)
W= rfftfreq(length, period)

cutoff = elzamalek_fft.copy()
ranges=[rang1,rang2,rang3]
ranges2=[rang1,rang2,(3000,8000)]
gain = -20000
for rang in ranges2:
    time = np.linspace(0,rang[1]-rang[0],len(cutoff[(W>rang[0])&(W<rang[1])]))
    han = 0.5-0.5*np.cos((2*np.pi*time)/(time[-1]-1))
    cutoff[(W>rang[0])&(W<rang[1])] *= 10**(-50* han /20)
    
plt.plot(W,abs(cutoff))
audio = irfft(cutoff)

```
