```python
import librosa
from scipy.fftpack import rfft, irfft, fftfreq, fft, rfftfreq
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt
from IPython.display import Audio
import numpy as np
```

```python
manganese, sr = librosa.load("audio/manganese.wav", duration=1.3)
f, t, Sxx = signal.spectrogram(manganese, sr)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
Audio(manganese, rate = sr)
```

```python
# seperate a from manganese
a, sr = librosa.load("audio/manganese.wav",offset=0.3 ,duration=0.172)
f, t, Sxx = signal.spectrogram(a, sr)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
Audio(manganese, rate = sr)
```

```python
#letter a forier analysis
period = 1/sr
length = len(a)
f_signal = rfft(a)
W= rfftfreq(length, period)
plt.plot(W, f_signal)
```
